```python
import pandas as pd
```

```python
df = pd.read_csv("16. covid_vaccine_statewise.csv")
df
```

```python
df.info()
```

```python
dic = {}
for i in range(7845):
    if df['State'][i] not in dic.keys():
        if df['First Dose Administered'][i] >= 0:
            dic[df['State'][i]] = df['First Dose Administered'][i]
        else:
            dic[df['State'][i]] = 0
    else:
        if df['First Dose Administered'][i] >= 0:
            dic[df['State'][i]] += df['First Dose Administered'][i]
dic = pd.DataFrame(dic.items())
cols = ["State", "Number of Individuals vaccinated for Dose 1"]
dic.columns = cols
dic
```

```python
dic1 = {}
for i in range(7845):
    if df['State'][i] not in dic1.keys():
        if df["Second Dose Administered"][i] >= 0:
            dic1[df["State"][i]] = df["Second Dose Administered"][i]
        else:
            dic1[df["State"][i]] = 0
    else:
        if df["Second Dose Administered"][i] >= 0:
            dic1[df["State"][i]] += df["Second Dose Administered"][i]
dic1 = pd.DataFrame(dic1.items())
cols = ["State", "Number of Individuals vaccinated for Dose 2"]
dic1.columns = cols
dic1
```
