```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("5-8-13-14.iris.csv")
df
```

```python
df.info()
```

```python
fig = plt.subplots(1, 1, figsize=(16,9))
plt.hist(df['sepal.length'])
```

```python
fig, axes = plt.subplots(1, 1, figsize=(16,9))
plt.hist(df['sepal.width'])
```

```python
fig, axes = plt.subplots(1, 1, figsize=(16,9))
plt.hist(df['petal.length'])
```

```python
fig, axes = plt.subplots(1, 1, figsize=(16,9))
plt.hist(df['petal.width'])
```

```python
fig, axes = plt.subplots(1, 1, figsize=(16,9))
axes.set_title("Distribution of Sepal Length")
sns.boxplot(y="sepal.length",x = "variety", data=df)
```

```python
fig, axes = plt.subplots(1, 1, figsize=(16,9))
axes.set_title("Distribution of Sepal Width")
sns.boxplot(y="sepal.width",x = "variety", data=df)
```

```python
fig, axes = plt.subplots(1, 1, figsize=(16,9))
axes.set_title("Distribution of Petal Length")
sns.boxplot(y="petal.length",x = "variety", data=df)
```

```python
fig, axes = plt.subplots(1, 1, figsize=(16,9))
axes.set_title("Distribution of Petal Width")
sns.boxplot(y="petal.width",x = "variety", data=df)
```
